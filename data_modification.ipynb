```python
import json
import os
import matplotlib.pyplot as plt  

folder_path = r'D:\UCL_codes\research project\Untitled Folder\Onefunction_PI'

save_path = 'train_data_strategy_1'
if not os.path.exists(save_path):
    os.makedirs(save_path)
    
    
for filename in os.listdir(folder_path):
    if filename.endswith('.json'):
        # 打开 JSON 文件
        with open(os.path.join(folder_path, filename), 'r') as f:
            data = json.load(f)

        xs = data['xs']
        ys = data['ys']
        x_obs = data['PI']['x_obs']
        y_obs = data['PI']['y_obs']
        max_value = max(ys)
        max_index = xs[ys.index(max_value)]
        
        print("最大值索引：", max_index, "最大值：", max_value, "obs最大值索引：", x_obs[y_obs.index(max(y_obs))], "obs最大值：", max(y_obs))

        couter = 0
        for i in range(0,len(y_obs)):
            if max_index-0.02 < x_obs[i] < max_index+0.02:
                print(i,couter)
                break
            else:
                couter +=1
                
        if couter==50:
            
            plt.plot(xs, ys, label = 'groundtruth')
            plt.scatter(x_obs, y_obs, label = 'observations',c='b')
            plt.legend()
            plt.title( 'PI data No.'+str(filename) )
            plt.show()
            
        y_obs.insert(0, 0)
        x_obs.append(0)
        rtg = list(range(couter, 0, -1)) + [0] * (52 - couter)
        r = [1] * couter + [0] * (52 - couter)
        if len(rtg)!=len(y_obs) or len(rtg)!=len(x_obs):
            print("error!!!", len(rtg), len(y_obs), len(x_obs))
        
        save_data = {'RTG':rtg, 'S':y_obs, 'A':x_obs, 'R':r}
        
        with open(os.path.join(save_path, filename), 'w') as f:
            json.dump(save_data, f)
        
        
```
